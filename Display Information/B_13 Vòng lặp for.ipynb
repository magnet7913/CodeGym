# Vòng lặp For

## For kết hợp với hàm range()

### Trường hợp 1 : Một tham số

```python
# in ra số tự nhiên từ 0-10
for i in range(11):
    print(i,end=' ')
```

```python
i = 0
while i <= 10:
    print(i,end=' ')
    i += 1
```

### ví dụ : in các số lẻ từ 0 -->100

```python
for i in range(101):
    if i%2==1:
        print(i,end=' ')
        
for i in range(1,101,2):
        print(i,end=' ')
```

```python
i=0
while i <= 100:
    if i%2==1:
        print(i,end=' ')
        i += 1
    else:
        i += 1
```

### Trường hợp 2 tham số 

```python
for i in range(0,100):
    if i%2==1:
        print(i,end=' ')
```

### ví dụ : in các số lẻ chia hết cho 7 từ a đến b

```python
a= int(input())
b= int(input())
for i in range(a,b):
    if i%7==1:
        print(i,end=' ')
```

```python
'''Thực Hành: Viết đoạn chương trình tính tích 1 * 2 * 3 *…* n với n được nhập từ bàn phím.'''
a = int(input())
b = 1
for a in range(1,a+1):
    b *= a
print(b)
```

### Trường hợp 3 tham số 

```python
for i in range(1,10,3):
    print(i,end=' ')
```

## Lặp qua một chuỗi 

```python
string = 'một hai ba bốn năm sáu'
for i in range(len(string)):
    print(f'value{i}, ký tự {string[i]}')
```

```python
string = 'một hai ba bốn năm sáu'
for i in string:
    print(f'{i}')
```

### Lặp qua một danh sách chuỗi

```python
name = ['một','hai','ba','bốn','năm','sáu']
for i in name:
    print(i)
```

## Vòng lặp lồng nhau

```python
# ví dụ : in ra bảng cửu chương
for i in range(1,11):
    for j in range(1,10):
        print(f'{i}x{j}={i*j}',end=' ')
```

```python
# ví dụ về hình lục giác bài turtle
import turtle
pen = turtle.Turtle()
pen.pensize(1)
pen.speed(0)
pen.color('black')

for i in range(10):
    for j in range(6):
        pen.forward(100)
        pen.right(60)
    pen.right(36)    
turtle.exitonclick()
```

```python
for i in range(1,4):
    for j in range(1,5):
        print(i*j)
```

## Câu lệnh pass

```python
string ='python'
for i in string:
    if i == 'h':
        pass
        print('Pass')
    print(i)
```

```python
password = 'abc'
login = False
while login==False:

    p = input('Nhập Password')
    if p == password:
        print('Đăng nhập thành công!')
        login = True
    else:
        print('Đăng nhập thất bại!')
```
