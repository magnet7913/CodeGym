```python
import turtle

background = turtle.Screen()
background.bgcolor("cyan")

pen = turtle.Turtle()
pen.pensize(3)
pen.speed(10)
pen.color('black')
pen.penup()
pen.goto(-500,-500)
pen.pendown()

#Khung nhà trước
pen.fillcolor("blue")
pen.begin_fill()
pen.forward(210)
pen.left(90)
pen.forward(210)
pen.left(90)
pen.forward(210)
pen.left(90)
pen.forward(210)
pen.end_fill()

#Khung cửa
pen.left(90)
pen.forward(70)
pen.fillcolor("light green")
pen.begin_fill()
pen.forward(70)
pen.left(90)
pen.forward(100)
pen.left(90)
pen.forward(70)
pen.left(90)
pen.forward(100)
pen.end_fill()

#Khung nhà bên hông
pen.penup()
pen.goto(-290,-500)
pen.pendown()
pen.fillcolor("yellow")
pen.begin_fill()
pen.goto(-150,-430)
pen.goto(-150,-220)
pen.goto(-290,-290)
pen.goto(-290,-500)
pen.end_fill()

#Cửa sổ bên hông
pen.penup()
pen.goto(-243,-371)
pen.pendown()
pen.fillcolor("red")
pen.begin_fill()
pen.goto(-196,-348)
pen.goto(-196,-298)
pen.goto(-243,-321)
pen.goto(-243,-371)
pen.end_fill()

#Mái nhà trước
pen.penup()
pen.goto(-500,-290)
pen.pendown()
pen.fillcolor("pink")
pen.begin_fill()
pen.goto(-395,-190)
pen.goto(-290,-290)
pen.goto(-500,-290)
pen.end_fill()

#Mái nhà bên hông
pen.penup()
pen.goto(-290,-290)
pen.pendown()
pen.fillcolor("orange")
pen.begin_fill()
pen.goto(-150,-220)
pen.goto(-220,-130)
pen.goto(-395,-190)
pen.goto(-290,-290)
pen.end_fill()

#Mặt trời
pen.penup()
pen.goto(290,290)
pen.pendown()
pen.fillcolor("yellow")
pen.begin_fill()
pen.circle(100,45)
pen.right(90)
pen.forward(45)
pen.backward(45)
pen.left(90)
pen.circle(100,45)
pen.right(90)
pen.forward(90)
pen.backward(90)
pen.left(90)
pen.circle(100,45)
pen.right(90)
pen.forward(45)
pen.backward(45)
pen.left(90)
pen.circle(100,45)
pen.right(90)
pen.forward(90)
pen.backward(90)
pen.left(90)
pen.circle(100,45)
pen.right(90)
pen.forward(45)
pen.backward(45)
pen.left(90)
pen.circle(100,45)
pen.right(90)
pen.forward(90)
pen.backward(90)
pen.left(90)
pen.circle(100,45)
pen.right(90)
pen.forward(45)
pen.backward(45)
pen.left(90)
pen.circle(100,45)
pen.right(90)
pen.forward(90)
pen.backward(90)
pen.left(90)
pen.end_fill()

#Thân cây
pen.penup()
pen.goto(290,-290)
pen.pendown()
pen.fillcolor("brown")
pen.begin_fill()
pen.goto(290,-390)
pen.goto(340,-390)
pen.goto(340,-290)
pen.goto(290,-290)
pen.end_fill()

#Lá cây
pen.fillcolor("green")
pen.begin_fill()
pen.goto(215,-290)
pen.goto(290,-215)
pen.goto(215,-215)
pen.goto(290,-140)
pen.goto(215,-140)
pen.goto(315,-55) #Đỉnh cây
pen.goto(415,-140)
pen.goto(340,-140)
pen.goto(415,-215)
pen.goto(340,-215)
pen.goto(415,-290)
pen.goto(290,-290)
pen.end_fill()

turtle.exitonclick()
```
