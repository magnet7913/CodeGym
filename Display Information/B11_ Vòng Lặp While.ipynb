## Sửa BTVN Tính phương trình bậc 2

```python
import math

# Nhập các hệ số a, b, c của phương trình bậc 2
a = float(input("Nhap a: "))
b = float(input("Nhap b: "))
c = float(input("Nhap c: "))

# Tính delta
delta = b*b - 4*a*c
if a == 0:
    if b == 0:
        print("phương trình có vô số nghiệm:")
    else:
        print(f"Phương trình về dạng bậc 1 và có nghiêm x = :{-c/b}")
# Tìm nghiệm của phương trình
else:
    if delta < 0:
      print("Phương trình vô nghiệm")
    elif delta == 0:
      x = -b / (2*a)
      print("Phương trình có nghiệm kép x =", x)
    else:
      x1 = (-b + math.sqrt(delta)) / (2*a)
      x2 = (-b - math.sqrt(delta)) / (2*a)
      print("Phương trình có 2 nghiệm phân biệt x1 =", x1, "va x2 =", x2)
```

# Vòng Lặp While

Định nghĩa

Trong Python vòng lặp while được sử dụng để thực thi một khối lệnh lặp đi lặp lại chừng nào điều kiện của vòng lặp vẫn còn đúng. Khi điều kiện sai, các lệnh sau vòng lặp while mới được thực hiện. Vòng lặp while thường được sử dụng trong các trường hợp không xác định được trước số vòng lặp.

## Cú pháp

### Vd: in ra tất cả các số chẵn từ 0 đên n

```python
n = int(input('Nhập số N bất kỳ: '))

even = 1
while even <= n :
    if even % 2 ==0:
        print(even,' ',end='')
    even += +1
```

### Thực Hành: In ra màn hình các số từ 0 đến 9

```python
num = 0
while num <= 9 :
    print(num,' ',end='')
    num += +1
```

### Thực hành : Tính tổng các số chia hết cho 3 trong khoảng từ 0 đến 100

```python
three = 0
total = 0
while three <= 100 :
    if three % 3 ==0:
        total += three
    three += 1
print(total)
```

### Thực hành : Nhập vào số nguyên dương n và in ra màn hình các số từ 0 đến n

```python
n = int(input('Nhập số N bất kỳ: '))

num = 0
while num <= n :
    print(num,' ',end='')
    num += 1
```

## Lệnh break

Lệnh break trong Python được sử dụng để điều khiển thoát khỏi vòng lặp khi thỏa mãn một điều kiện bên ngoài nào đó. Lệnh break được đặt ở trong vòng lặp

### Cú pháp

break</p>

### Lưu ý ***
#### lệnh break bắt buộc phải đặt trong vòng lặp
#### lệnh break sẽ hủy vòng lặp gần nhất (trong trường hợp có nhiều vòng lặp lồng nhau)

vd : Nhập vào số nguyên dương n và in ra màn hình các số từ 0 đến n, nếu nhập vào số âm thì dừng vòng lặp

```python
# Khởi tạo biến i = 0
i = 0

# Vòng lặp while
while True:
   # Nhập vào số nguyên dương n
   n = int(input("Nhập vào số nguyên dương n: "))

   # Nếu n là số âm, dừng vòng lặp
   if n < 0:
      break

   # In ra màn hình các số từ 0 đến n
   while i <= n:
      print(i)
      i += 1
```

### Thực hành : In ra màn hình các số từ 1 đến 10, nếu nhập vào số lớn hơn 10 thì dừng vòng lặp

```python
i=1
while True:
    n = int(input('Nhập vào một số bất kỳ: '))
    if n > 10:
        break
    while i <= n:
        print(i,' ',end='')
        i += 1
    break
```

## Lệnh continue

Lệnh continue cũng là một lệnh điều khiển vòng lặp giống như break. Chỉ khác ở chỗ, continue thay vì thoát ra khỏi vòng lặp như break thì continue bỏ qua vòng lặp hiện tại và tiến hành thực thi ở bước lặp kế tiếp.

### Cú Pháp

continue

Ví dụ: In ra màn hình các số chia hết cho 2 trong khoảng từ 0 đến 10

```python
# Khởi tạo biến i = 0
i = 0

# Vòng lặp while
while i <= 10:
   # Nếu i không chia hết cho 2, bỏ qua số đó và lặp lại vòng lặp
   if i % 2 != 0:
      i += 1
      continue

   # In ra màn hình số i nếu nó chia hết cho 2
   print(i)
   i += 1
```

### Thực Hành : In ra màn hình các số lẻ trong khoảng từ 1 đến 10

```python
i = 0
while i <= 10:
    if i % 2 != 1:
        i += 1
        continue
    print(i,' ',end='')
    i += 1
```

### Thực hành : Tính tổng các số chia hết cho 3 trong khoảng từ 0 đến 100

```python
three = 0
total = 0
while three <= 100 :
    if three % 3 !=0:
        three += 1
        continue
    total += three
    three += 1
print(total)
```
