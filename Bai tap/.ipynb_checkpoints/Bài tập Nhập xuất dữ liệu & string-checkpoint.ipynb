```python
usd_amount = int(input('Nhập số tiền USD cần đổi: '))
usd_vnd = int(input('Nhập tỷ giá USD/VND: '))
vnd_final = usd_amount * usd_vnd
print(f"Bạn sẽ đổi {usd_amount} với tỷ giá 1 USD bằng {usd_vnd} VND để lấy {vnd_final} VND.")
```

```python
house_lox = int(input("Nhập vị trí đặt nhà trên trục x: "))
house_loy = int(input("Nhập vị trí đặt nhà trên trục y: "))
house_x = int(input("Bề dài căn nhà: "))
house_y = int(input("Bề ngang căn nhà: "))
house_col = input("Màu tường nhà: ")
door_col = input("Màu cửa nhà: ")
roof_col = input("Màu mái nhà: ")
tree1_lox = int(input("Nhập vị trí đặt cây 1 trên trục x: "))
tree1_loy = int(input("Nhập vị trí đặt cây 1 trên trục y: "))
tree1_x = int(input("Nhập bề ngang tán cây 1: "))
tree2_lox = int(input("Nhập vị trí đặt cây 2 trên trục x: "))
tree2_loy = int(input("Nhập vị trí đặt cây 2 trên trục y: "))
tree2_x = int(input("Nhập bề ngang tán cây 2: "))
```

```python

import turtle as pen

background = pen.Screen()

pen.pensize(3)
pen.speed(0)
pen.color('blue')
pen.up()
pen.goto(house_lox,house_loy)
pen.down()

# Khung nhà
pen.fillcolor(house_col)
pen.begin_fill()
pen.forward(house_x/5*3)
pen.left(90)
pen.forward(house_y)
pen.left(90)
pen.forward(house_x)
pen.left(90)
pen.forward(house_y)
pen.left(90)
pen.forward(house_x/5*2)
pen.end_fill()

# Cửa nhà
pen.fillcolor(door_col)
pen.begin_fill()
pen.left(90)
pen.forward(house_y/3*2)
pen.right(90)
pen.forward(house_x/5)
pen.right(90)
pen.forward(house_y/3*2)
pen.end_fill()

# Mái nhà
pen.up()
pen.goto(house_lox - 2*house_x/5,house_loy + house_y)
pen.down()
pen.fillcolor(roof_col)
pen.begin_fill()
pen.goto(house_lox-0.5*house_x/5,house_loy+1.5*house_y)
pen.goto(house_lox+house_x/5,house_loy+house_y)
pen.end_fill()

# Ống khói
pen.up()
pen.goto(house_lox,house_loy+1.333*house_y)
pen.down()
pen.fillcolor(roof_col)
pen.begin_fill()
pen.goto(house_lox,house_loy+1.666*house_y) #đỉnh trái
pen.goto(house_lox+0.3*house_x/5,house_loy+1.666*house_y)#đỉnh phải
pen.goto(house_lox+0.3*house_x/5,house_loy+1.24*house_y)
pen.end_fill()

# Khói
pen.up()
pen.goto(house_lox+0.05*house_x/5,house_loy+1.75*house_y)
pen.down()
pen.goto(house_lox+0.05*house_x/5,house_loy+2*house_y)
pen.up()
pen.goto(house_lox+0.2995*house_x/5,house_loy+1.7*house_y)
pen.down()
pen.goto(house_lox+0.2995*house_x/5,house_loy+2.1*house_y)

# Cây 1
#Lá cây
pen.up()
pen.goto(tree1_lox,tree1_loy)
pen.down()
pen.fillcolor("green")
pen.begin_fill()
#Tán 1
pen.goto(tree1_lox+tree1_x/2,tree1_loy)
pen.goto(tree1_lox,tree1_loy+tree1_x/3*2)
pen.goto(tree1_lox-tree1_x/2,tree1_loy)
pen.goto(tree1_lox,tree1_loy)

#Tán 2
pen.up()
pen.goto(tree1_lox,tree1_loy+tree1_x/3*2)
pen.down()
pen.goto(tree1_lox+0.9*tree1_x/2,tree1_loy+tree1_x/3*2)
pen.goto(tree1_lox,tree1_loy+1.9*tree1_x/3*2)
pen.goto(tree1_lox-0.9*tree1_x/2,tree1_loy+tree1_x/3*2)
pen.goto(tree1_lox,tree1_loy+tree1_x/3*2)

#Tán 3
pen.up()
pen.goto(tree1_lox,tree1_loy+1.9*tree1_x/3*2)
pen.down()
pen.goto(tree1_lox+0.81*tree1_x/2,tree1_loy+1.9*tree1_x/3*2)
pen.goto(tree1_lox,tree1_loy+2.71*tree1_x/3*2)
pen.goto(tree1_lox-0.81*tree1_x/2,tree1_loy+1.9*tree1_x/3*2)
pen.goto(tree1_lox,tree1_loy+1.9*tree1_x/3*2)
pen.end_fill()

#Thân cây
pen.up()
pen.goto(tree1_lox-0.2*tree1_x/2,tree1_loy)
pen.down()
pen.fillcolor("brown")
pen.begin_fill()
pen.goto(tree1_lox-0.35*tree1_x/2,tree1_loy-tree1_x/2)
pen.goto(tree1_lox+0.35*tree1_x/2,tree1_loy-tree1_x/2)
pen.goto(tree1_lox+0.2*tree1_x/2,tree1_loy)
pen.end_fill()

# Cây 2
#Lá cây
pen.up()
pen.goto(tree2_lox,tree2_loy)
pen.down()
pen.fillcolor("green")
pen.begin_fill()
#Tán 1
pen.goto(tree2_lox+tree2_x/2,tree2_loy)
pen.goto(tree2_lox,tree2_loy+tree2_x/3*2)
pen.goto(tree2_lox-tree2_x/2,tree2_loy)
pen.goto(tree2_lox,tree2_loy)

#Tán 2
pen.up()
pen.goto(tree2_lox,tree2_loy+tree2_x/3*2)
pen.down()
pen.goto(tree2_lox+0.9*tree2_x/2,tree2_loy+tree2_x/3*2)
pen.goto(tree2_lox,tree2_loy+1.9*tree2_x/3*2)
pen.goto(tree2_lox-0.9*tree2_x/2,tree2_loy+tree2_x/3*2)
pen.goto(tree2_lox,tree2_loy+tree2_x/3*2)

#Tán 3
pen.up()
pen.goto(tree2_lox,tree2_loy+1.9*tree2_x/3*2)
pen.down()
pen.goto(tree2_lox+0.81*tree2_x/2,tree2_loy+1.9*tree2_x/3*2)
pen.goto(tree2_lox,tree2_loy+2.71*tree2_x/3*2)
pen.goto(tree2_lox-0.81*tree2_x/2,tree2_loy+1.9*tree2_x/3*2)
pen.goto(tree2_lox,tree2_loy+1.9*tree2_x/3*2)
pen.end_fill()

#Thân cây
pen.up()
pen.goto(tree2_lox-0.2*tree2_x/2,tree2_loy)
pen.down()
pen.fillcolor("brown")
pen.begin_fill()
pen.goto(tree2_lox-0.35*tree2_x/2,tree2_loy-tree2_x/2)
pen.goto(tree2_lox+0.35*tree2_x/2,tree2_loy-tree2_x/2)
pen.goto(tree2_lox+0.2*tree2_x/2,tree2_loy)
pen.end_fill()



pen.done()
```
